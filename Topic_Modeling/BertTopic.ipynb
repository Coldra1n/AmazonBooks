```python
import pandas as pd
from bertopic import BERTopic

# Load the dataset
book_reviews_path = 'C:/Users/Nick/Desktop/Amazon_prediction/data/book_reviews_2023-12-12-12.csv'
book_reviews = pd.read_csv(book_reviews_path)

# Ensure that the review text is in string format
book_reviews['review_text'] = book_reviews['review_text'].astype(str)

# Initialize BERTopic
topic_model = BERTopic(language="english", calculate_probabilities=True)

# Fit the model on your review texts
topics, probabilities = topic_model.fit_transform(book_reviews['review_text'])

# To explore the topics, you can use
topic_info = topic_model.get_topic_info()
print(topic_info.head(10))  # Display top 10 topics

# To see the top words in a specific topic
top_words = topic_model.get_topic(0)  # Replace 0 with your topic number
print(top_words)
```
